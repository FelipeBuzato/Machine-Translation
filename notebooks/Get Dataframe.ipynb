```python
import pandas as pd

# Lê apenas as duas primeiras colunas
df = pd.read_csv(
    "por.txt",
    sep="\t",
    header=None,
    usecols=[0, 1],
    names=["english", "portuguese"],
    encoding="utf-8"
)

# Remove frases vazias
df.dropna(inplace=True)

# Remove duplicatas
df.drop_duplicates(inplace=True)

# Calcula o número de palavras
df["en_len"] = df["english"].str.split().str.len()
df["pt_len"] = df["portuguese"].str.split().str.len()

# Mantém apenas frases entre 2 e 30 palavras
df = df[
    (df["en_len"] >= 2) &
    (df["en_len"] <= 30) &
    (df["pt_len"] >= 2) &
    (df["pt_len"] <= 30)
]

# Remove as colunas auxiliares
df.drop(columns=["en_len", "pt_len"], inplace=True)
```

```python
# Embaralha
df = df.sample(frac=1).reset_index(drop=True)

df.iloc[:50, :]
```

```python
# Seleciona 29 mil exemplos
df_train = df.iloc[:29000]
df_validation = df.iloc[29000:30000]
df_test = df.iloc[30000:40000]

# Salva
df_train.to_csv(
    "train.csv",
    index=False,
    encoding="utf-8-sig"
)

# Salva
df_validation.to_csv(
    "validation.csv",
    index=False,
    encoding="utf-8-sig"
)

# Salva
df_test.to_csv(
    "test.csv",
    index=False,
    encoding="utf-8-sig"
)

print(df_train.head())
print(f"\nNúmero de exemplos: {len(df_train)}, {len(df_validation)}, {len(df_test)}")
```
